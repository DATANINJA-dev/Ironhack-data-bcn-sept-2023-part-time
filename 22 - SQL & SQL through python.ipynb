# Subqueries

Subqueries are nested queries within other queries. A subquery returns results that can be used by the main query in which it is embedded, further transformed, or joined with other tables or subqueries. Let's look at an example using our publications dataset.

## What are all the authors_id of those authors who published - `where` & `from`

````python
def query_autores_subquery (filtro, tabla_filtro, otra_tabla):
    dbName = "publications"
    connectionData=f"mysql+pymysql://root:{password}@localhost/{dbName}"
    engine = alch.create_engine(connectionData)
    
    subquery = f"(select {filtro} from {tabla_filtro})"
    
    query_publicados = f"""
        SELECT * FROM {otra_tabla}
        WHERE {filtro} in
            {subquery};
        """
    
    df = pd.read_sql_query (query_publicados, engine)
    
    return df
````

## What's the average selling of unique items & the average quantity of units sold per store? -  `from`

Suppose we want to created a query to summarize the store's sales, including the number of orders, the number of items, and the total number of units sold.

Simple query (not subquery). Store sales including:
- number of orders
- number of items (count item)
- total number of units sold (sum Qty)

````python
# USING JOIN

# 1. Tables: stores & sales
# 2. Select: store_id (stores), qty (sales), title_id(sales), ord_num (sales) 
# 3. Cómo: 


pd.read_sql_query("""
SELECT stores.stor_name AS store_name,
        COUNT(DISTINCT(ord_num)) AS Orders,
        COUNT(DISTINCT(sales.title_id)) as Items,
        SUM(qty) as Quantity

FROM stores
    JOIN sales
    ON stores.stor_id = sales.stor_id
    
-- GROUP BY 1;

GROUP BY stores.stor_id
ORDER BY Quantity DESC;

""",engine)
````

If we wanted to display the average number of items per order and the average quantity per item for each store, we could do so by embedding this query as a subquery in another query that performs those calculations.

````python
"""

-- name, item, order, qty
-- # 1. STORE, AVGITEMS, AVGQTY
-- # 2. tabla: 

SELECT store_name, Items/Orders as avgItems, Quantity/Orders as avgQuantity
FROM

    (SELECT stores.stor_name AS store_name,
            COUNT(DISTINCT(ord_num)) AS Orders,
            COUNT(DISTINCT(sales.title_id)) as Items,
            SUM(qty) as Quantity

    FROM stores
        JOIN sales
        ON stores.stor_id = sales.stor_id

    -- GROUP BY 1;

    GROUP BY stores.stor_id
    ORDER BY Quantity DESC)
;
"""
````

````python
pd.read_sql_query("""

SELECT store_name, Items/Orders as avgItems, Quantity/Orders as avgQuantity
FROM

    (SELECT stores.stor_name AS store_name,
            COUNT(DISTINCT(ord_num)) AS Orders,
            COUNT(DISTINCT(sales.title_id)) as Items,
            SUM(qty) as Quantity

    FROM stores
        JOIN sales
        ON stores.stor_id = sales.stor_id

    GROUP BY stores.stor_id
    ORDER BY Quantity DESC)  AS resultado_de_la_subquery

""", engine)
````

The main query uses the results of the subquery and allows us to perform the calculations using the field names that we assign in the subquery. To access the results of the subquery, we also need to enclose it in parentheses and give it a name.

A main query can contain multiple subqueries, sometimes with other subqueries nested within them as well. From the perspective of the main query, the results returned by a subquery serve the same purpose as a table. They are a data source for the main query. In fact, you can even join tables and results of subqueries within the main query.

## Temporary Tables

Temporary tables are staging tables that are built from the results of queries and dropped after the database session has ended. They serve to store the results of the query for later use in the session. The way to avoid having to write the same subquery multiple times is to save the query results to a temporary table and then query the temporary table each time we need to access those results.

Let's see how we could have changed the subquery from the previous section to a temporary table. To do this, we can use the CREATE TEMPORARY TABLE command, followed by the name we want to give the new table and the query.

````python
"""CREATE TEMPORARY TABLE publications.store_sales_summary

SELECT stores.stor_name AS store_name,
        COUNT(DISTINCT(ord_num)) AS Orders,
        COUNT(DISTINCT(sales.title_id)) as Items,
        SUM(qty) as Quantity

FROM stores
    JOIN sales
    ON stores.stor_id = sales.stor_id
    
GROUP BY stores.stor_id
ORDER BY Quantity DESC;
"""
````
